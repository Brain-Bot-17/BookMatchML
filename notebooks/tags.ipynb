```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
tags = pd.read_csv(r"C:/Users/HP/OneDrive\Desktop\BookMatchML\data/tags.csv")

# Function to analyze data
def analyze_data(df, stage="Before Cleaning"):
    print(f"\n--- {stage} ---")
    print("Data Info:")
    print(df.info())
    print("\nNumeric Summary:")
    print(df.describe())
    print("\nCategorical Summary:")
    if any(df.dtypes == 'object'):
        print(df.describe(include='object'))
    else:
        print("No categorical columns present.")

    # Plots
    plt.figure(figsize=(15, 5))
    # Numeric: tag_id
    plt.subplot(1, 2, 1)
    sns.histplot(df['tag_id'].dropna(), bins=50, kde=True)
    plt.title(f'Tag ID Distribution ({stage})')
    plt.xlabel('Tag ID')
    plt.ylabel('Count')
    
    # Categorical: tag_name value counts (top 10)
    plt.subplot(1, 2, 2)
    top_tags = df['tag_name'].value_counts().head(10)
    sns.barplot(x=top_tags.index, y=top_tags.values)
    plt.title(f'Top 10 Tag Names ({stage})')
    plt.xlabel('Tag Name')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Before Cleaning Analysis
analyze_data(tags, "Before Cleaning")

# Additional Checks
print("\nAdditional Checks:")
print("Missing values per column:")
print(tags.isnull().sum())
print("Duplicate rows:", tags.duplicated().sum())
print("Duplicate tag_ids:", tags['tag_id'].duplicated().sum())
print("Duplicate tag_names:", tags['tag_name'].duplicated().sum())
print("Unique tag_ids:", tags['tag_id'].nunique())
print("Unique tag_names:", tags['tag_name'].nunique())
print("Sample tag_names:", tags['tag_name'].head(10).tolist())

# Load book_tags.csv for comparison
book_tags = pd.read_csv(r"C:/Users/HP/OneDrive\Desktop\BookMatchML\data\book_tags_cleaned.csv")
print("Book_tags.csv unique tag_ids:", book_tags['tag_id'].nunique())
print("Overlap with book_tags.csv tag_ids:", len(set(tags['tag_id']).intersection(set(book_tags['tag_id']))))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import unicodedata

# Load datasets
tags = pd.read_csv(r"C:/Users/HP/OneDrive\Desktop\BookMatchML\data\tags.csv")
book_tags = pd.read_csv(r"C:/Users/HP/OneDrive\Desktop\BookMatchML\data\book_tags_cleaned.csv")

# Function to analyze data
def analyze_data(df, stage="Before Cleaning"):
    print(f"\n--- {stage} ---")
    print("Data Info:")
    print(df.info())
    print("\nNumeric Summary:")
    print(df.describe())
    print("\nCategorical Summary:")
    if any(df.dtypes == 'object'):
        print(df.describe(include='object'))
    else:
        print("No categorical columns present.")

    # Plots
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 2, 1)
    sns.histplot(df['tag_id'].dropna(), bins=50, kde=True)
    plt.title(f'Tag ID Distribution ({stage})')
    plt.subplot(1, 2, 2)
    top_tags = df['tag_name'].value_counts().head(10)
    sns.barplot(x=top_tags.index, y=top_tags.values)
    plt.title(f'Top 10 Tag Names ({stage})')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Before Cleaning Analysis
analyze_data(tags, "Before Cleaning")

# Additional Checks
print("\nAdditional Checks (Before):")
print("Missing values per column:")
print(tags.isnull().sum())
print("Duplicate tag_ids:", tags['tag_id'].duplicated().sum())
print("Duplicate tag_names:", tags['tag_name'].duplicated().sum())
print("Unique tag_ids:", tags['tag_id'].nunique())
print("Unique tag_names:", tags['tag_name'].nunique())
print("Sample tag_names (first 10):", tags['tag_name'].head(10).tolist())
print("Sample tag_names (random 10):", tags['tag_name'].sample(10, random_state=42).tolist())

# Preprocessing
tags_cleaned = tags.copy()

# 1. Normalize tag_names
tags_cleaned['tag_name'] = tags_cleaned['tag_name'].apply(lambda x: unicodedata.normalize('NFKC', x).strip())
print("Normalized fullwidth characters in tag_names.")

# 2. Identify and optionally filter meaningless tag_names
placeholder_tags = tags_cleaned['tag_name'].str.match(r'^[-–—]+$|^--?\d+-?$|^[0-9]+$')
if placeholder_tags.sum() > 0:
    print(f"Found {placeholder_tags.sum()} potentially meaningless tag_names.")
    print("Sample meaningless tag_names:", tags_cleaned[placeholder_tags]['tag_name'].head().tolist())
    # Optional: Uncomment to remove meaningless tags
    # tags_cleaned = tags_cleaned[~placeholder_tags]
    # print("Removed meaningless tag_names.")

# 3. Validate tag_id
valid_tag_ids = set(book_tags['tag_id'])
invalid_tag_ids = tags_cleaned[~tags_cleaned['tag_id'].isin(valid_tag_ids)]['tag_id'].unique()
if len(invalid_tag_ids) > 0:
    print(f"Found {len(invalid_tag_ids)} invalid tag_ids:", sorted(invalid_tag_ids)[:10])
    tags_cleaned = tags_cleaned[tags_cleaned['tag_id'].isin(valid_tag_ids)]
    print("Removed tags with invalid tag_ids.")
else:
    print("All tag_ids match book_tags.csv.")

# After Cleaning Analysis
analyze_data(tags_cleaned, "After Cleaning")

# Final Validation
print("\nFinal Validation:")
print("Missing values:", tags_cleaned.isnull().sum().sum())
print("Duplicate tag_ids:", tags_cleaned['tag_id'].duplicated().sum())
print("Unique tag_ids:", tags_cleaned['tag_id'].nunique())
print("Unique tag_names:", tags_cleaned['tag_name'].nunique())
print("Rows remaining:", len(tags_cleaned))
print("Overlap with book_tags.csv tag_ids:", len(set(tags_cleaned['tag_id']).intersection(set(book_tags['tag_id']))))

# Save cleaned data
tags_cleaned.to_csv(r"C:/Users/HP/OneDrive\Desktop\BookMatchML\data/tags_cleaned.csv", index=False)
print("Cleaned data saved to 'tags_cleaned.csv'")
```
