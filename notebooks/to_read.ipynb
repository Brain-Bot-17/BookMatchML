```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
to_read = pd.read_csv(r"C:/Users/HP/OneDrive\Desktop\BookMatchML\data\to_read.csv")
ratings = pd.read_csv(r"C:/Users/HP/OneDrive\Desktop\BookMatchML\data\ratings_cleaned.csv")

# Function to analyze data
def analyze_data(df, stage="Before Cleaning"):
    print(f"\n--- {stage} ---")
    print("Data Info:")
    print(df.info())
    print("\nNumeric Summary:")
    print(df.describe())
    print("\nCategorical Summary:")
    if any(df.dtypes == 'object'):
        print(df.describe(include='object'))
    else:
        print("No categorical columns present.")

    # Plots
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 2, 1)
    sns.histplot(df['user_id'].dropna(), bins=50, kde=True)
    plt.title(f'User ID Distribution ({stage})')
    plt.subplot(1, 2, 2)
    sns.histplot(df['book_id'].dropna(), bins=50, kde=True)
    plt.title(f'Book ID Distribution ({stage})')
    plt.tight_layout()
    plt.show()

# Before Cleaning Analysis
analyze_data(to_read, "Before Cleaning")

# Additional Checks
print("\nAdditional Checks (Before):")
print("Missing values per column:")
print(to_read.isnull().sum())
print("Duplicate rows:", to_read.duplicated().sum())
print("Duplicate user-book pairs:", to_read.duplicated(subset=['user_id', 'book_id']).sum())
print("Unique users:", to_read['user_id'].nunique())
print("Unique books:", to_read['book_id'].nunique())
print("Average books per user:", to_read.groupby('user_id')['book_id'].count().mean())

# Preprocessing
to_read_cleaned = to_read.copy()

# 1. Validate user_id and book_id against ratings.csv
valid_user_ids = set(ratings['user_id'])
valid_book_ids = set(ratings['book_id'])
invalid_users = to_read_cleaned[~to_read_cleaned['user_id'].isin(valid_user_ids)]['user_id'].unique()
invalid_books = to_read_cleaned[~to_read_cleaned['book_id'].isin(valid_book_ids)]['book_id'].unique()

if len(invalid_users) > 0:
    print(f"Found {len(invalid_users)} invalid user_ids:", sorted(invalid_users)[:10])
    to_read_cleaned = to_read_cleaned[to_read_cleaned['user_id'].isin(valid_user_ids)]
    print("Removed rows with invalid user_ids.")
else:
    print("All user_ids match ratings.csv.")

if len(invalid_books) > 0:
    print(f"Found {len(invalid_books)} invalid book_ids:", sorted(invalid_books)[:10])
    to_read_cleaned = to_read_cleaned[to_read_cleaned['book_id'].isin(valid_book_ids)]
    print("Removed rows with invalid book_ids.")
else:
    print("All book_ids match ratings.csv.")

# After Cleaning Analysis
analyze_data(to_read_cleaned, "After Cleaning")

# Final Validation
print("\nFinal Validation:")
print("Missing values:", to_read_cleaned.isnull().sum().sum())
print("Duplicate user-book pairs:", to_read_cleaned.duplicated(subset=['user_id', 'book_id']).sum())
print("Unique users:", to_read_cleaned['user_id'].nunique())
print("Unique books:", to_read_cleaned['book_id'].nunique())
print("Rows remaining:", len(to_read_cleaned))
print("Overlap with ratings.csv user_ids:", len(set(to_read_cleaned['user_id']).intersection(set(ratings['user_id']))))
print("Overlap with ratings.csv book_ids:", len(set(to_read_cleaned['book_id']).intersection(set(ratings['book_id']))))

# Save cleaned data
to_read_cleaned.to_csv(r"C:/Users/HP/OneDrive\Desktop\BookMatchML\data\to_read_cleaned.csv", index=False)
print("Cleaned data saved to 'to_read_cleaned.csv'")
```
